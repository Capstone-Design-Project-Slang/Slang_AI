```python
import numpy as np
import os

os.environ['CUDA_VISIBLE_DEVICES'] = '1'
os.environ['TF_FORCE_GPU_ALLOW_GROWTH'] = 'true'
```

```python
actions = ['축구','야구', '농구', '배구', '탁구', '테니스', '태권도', '씨름', '유도', '수영', '스케이트', '스키',  '사격', '팬싱', '검도']
```

```python
lstdir = os.listdir("dataset")
len_actions = len(lstdir)
data_lst = []

print(lstdir)

for i in range(len_actions):
    path_temp = os.path.join("dataset", os.listdir("dataset")[i])
    data_temp = np.load(path_temp)

    if data_temp.shape[0]:
        data_lst.append(data_temp)

print(data_lst)
```

```python
data = np.concatenate(data_lst)
data.shape
```

```python
X = data[:, :-1]
y = data[:, -1]

print(f"dim(X) = {X.shape}\ndim(y) = {y.shape}")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, stratify=y, random_state=17)
print(f"dim(X_train) = {X_train.shape}\ndim(X_test) = {X_test.shape}\ndim(y_train) = {y_train.shape}\ndim(y_test) = {y_test.shape}")
```

```python
import sys
sys.executable
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import GridSearchCV

# 일단은 튜닝 없는 모델로 실행가능성만 테스트하는 단계
# 이후 demo version에서 튜닝 프로세스 추가할 예정!

# param_grid = {
#     'num_leaves': [31, 63,127],
#     'learning_rate': [0.1, 0.01],
#     'n_estimators': [50, 100,1000],
#     'min_child_samples': [10,20,30],
# }
# scoring = 'f1_macro'

lgbm = LGBMClassifier()
lgbm.fit(X_train, y_train)
```

```python
from sklearn.metrics import f1_score

y_pred = lgbm.predict(X_test)
f1_score(y_test, y_pred, average="macro")
```
